## Prerequisites Setup

Here is a **detailed, beginner-friendly explanation** of the CUDA fundamentals you must know before learning TensorRT. Since you already know AI/ML, I’ll explain everything with a focus on **deep learning inference optimization**.

────────────────────────────
CUDA Basics (Foundation for TensorRT)

What is CUDA?
CUDA (Compute Unified Device Architecture) is NVIDIA’s parallel computing platform that lets you run code directly on the GPU. Deep learning frameworks (PyTorch, TensorFlow), optimization frameworks (TensorRT), and inference servers (Triton) all use CUDA kernels internally.

The GPU has a design that allows massive parallelism—thousands of operations happen simultaneously, unlike CPUs which are optimized for sequential logic.

────────────────────────────

1. GPU Architecture (How Computation Happens on GPU)

To understand TensorRT optimizations, you must know how a GPU is structured.

The GPU architecture has these major components:

1.1 SM (Streaming Multiprocessor)
An SM is like a “mini CPU” block inside the GPU.
A modern GPU often has **20–128+ SMs** depending on the model.

Each SM contains:
• CUDA cores (for FP32)
• Tensor Cores (for FP16/INT8 matrix math)
• Registers
• Shared memory
• Warp schedulers

TensorRT tries to use SMs as efficiently as possible.

1.2 CUDA Cores
These are basic execution units.
They do simple operations:
• Add
• Multiply
• Activation functions
• Element-wise operations

CUDA cores execute **FP32 (single-precision)** operations.

Example:
ResNet50 in FP32 uses mostly CUDA cores → slower compared to FP16/INT8.

1.3 Tensor Cores
Tensor Cores are specialized units inside SMs designed for **matrix multiplication**, the core operation of deep learning.

They accelerate:
• MatMul
• Convolutions (im2col → MatMul)
• Transformers (QKV projections, attention)

Tensor Cores process:
• FP16
• BF16
• INT8
• FP8 (on newer GPUs)

TensorRT uses Tensor Cores heavily → you get 2× to 8× faster inference.

1.4 Memory Hierarchy
The GPU memory system is critical for performance.

Global Memory (VRAM) – largest, slowest
Shared Memory – small, very fast
L2 Cache – medium
Registers – ultra fast

TensorRT optimizes memory usage by:
• kernel fusion
• reducing memory transfers
• using shared memory for intermediate results

This is why TensorRT is so fast.

────────────────────────────
2. CUDA Execution Model (Warps & Threads)

When software runs on a GPU, CUDA launches thousands of threads.

Key concepts:

2.1 Thread
Smallest compute unit executed on GPU.

2.2 Block
Group of threads.

2.3 Grid
Collection of blocks.

2.4 Warp
A group of 32 threads that execute the same instructions together.

Deep learning kernels (matmul, conv) are executed as warps.

TensorRT optimizes by selecting kernels that maximize warp efficiency.

Example:
If a model has an odd input shape like 255×13×13, warps may be underutilized → slower inference.
TensorRT fixes this by auto-tuning the best kernel tactic.

────────────────────────────
3. Numeric Precisions (FP32 → FP16 → INT8)

TensorRT’s biggest speedups come from lower-precision inference.

Let’s break down each precision format.

3.1 FP32 (Single Precision)

FP32 has 32 bits:
• 1 sign bit
• 8-bit exponent
• 23-bit fraction

Pros
✓ High accuracy
✓ Stable training

Cons
✗ Slowest inference
✗ High GPU memory usage
✗ Cannot use Tensor Cores fully

Used for:
• Model training
• Sensitive inference models

TensorRT rarely uses pure FP32 unless necessary.

3.2 FP16 (Half Precision)

FP16 = 16 bits
• 1 bit sign
• 5-bit exponent
• 10-bit mantissa

Pros
✓ 2× faster than FP32
✓ Uses Tensor Cores → BIG speedup
✓ Half memory usage
✓ Good accuracy for inference

Cons
• Slight precision loss
• Some layers may fall back to FP32

Used for:
• Almost all TensorRT deployment
• CNNs, Transformers

On RTX GPUs, FP16 offers **2–4×** throughput boost.

3.3 INT8 (Integer Precision)

INT8 = 8-bit integers representing values scaled by a factor.

Example:
Real value ≈ int_value × scale factor

Pros
✓ Fastest precision
✓ 4× smaller than FP32
✓ 2–8× faster with Tensor Cores
✓ Great for edge devices (Jetson)

Cons
• Needs calibration dataset
• May reduce accuracy slightly
• Some layers can’t use INT8

TensorRT uses INT8 quantization with:
• Entropy calibrator
• MinMax calibrator
• PTQ (Post Training Quantization)
• Accuracy-aware quantization

INT8 is the key for **real-time**, **low-latency**, **high-throughput** inference.

────────────────────────────
4. How TensorRT Uses These Concepts

Knowing CUDA basics helps understand TensorRT optimizations:

• TensorRT fuses layers to reduce global memory trips
• Chooses best kernel tactics to maximize warp usage
• Uses Tensor Cores automatically for FP16 & INT8
• Allocates memory efficiently for SMs
• Minimizes precision loss while maximizing speed

Example Speed Comparison (Typical GPU)
FP32 → 100 ms inference
FP16 → 35 ms
INT8 → 15 ms

### what exatly is fp16,gp16 and int8

Here is a **simple, clear explanation of FP16, BF16, FP32, and INT8**—what they actually *are* and *why* they matter in AI/ML and TensorRT.

──────────────────────────
What Exactly Are FP32, FP16, BF16, and INT8?

These are **numeric formats** used to store and compute numbers inside neural networks.

In deep learning, everything — weights, activations, gradients — is basically large matrices of numbers.
These numbers are stored in a specific format with a fixed number of bits.

Different formats change:
• how much memory they use
• how fast GPU can process them
• how accurate calculations are

Let's break them down one by one.

──────────────────────────

1. FP32 (Single-Precision Float)

FP32 is a **32-bit floating-point number**.

Bits:
• 1 sign bit
• 8 exponent bits
• 23 mantissa bits (precision)

What it means:
✓ Very accurate
✓ Large range
✓ Best for training

But:
✗ Slowest
✗ Uses most memory
✗ Cannot fully utilize Tensor Cores

Example:
Number: 1.23456789
Stored with very high precision.

──────────────────────────
2. FP16 (Half-Precision Float)

FP16 is a **16-bit floating-point number**.

Bits:
• 1 sign bit
• 5 exponent bits
• 10 mantissa bits

Compared to FP32:
• Half memory
• Faster computation
• Slightly lower precision

Advantages:
✓ Much faster (Tensor Cores love FP16)
✓ Lower VRAM usage
✓ Great for inference

Disadvantages:
• Smaller dynamic range
• Some numbers underflow/overflow

Example:
1.23456789 → 1.2346 (slight rounding)

──────────────────────────
3. BF16 (Bfloat16)

BF16 = **Brain Floating Point 16**.
It is also 16 bits, but laid out differently than FP16.

Bits:
• 1 sign bit
• 8 exponent bits → **same as FP32**
• 7 mantissa bits

Why it exists:
• FP32-like range
• FP16-like memory size
• Less precision than FP16

Used mainly for training on TPUs, new NVIDIA GPUs, PyTorch AMP.

FP16 = high precision but small range
BF16 = low precision but large range

──────────────────────────
4. INT8 (8-bit Integer)

INT8 is **not floating-point**.
It stores numbers as 8-bit integers (0 to 255 or -128 to +127).

INT8 isn't accurate enough for raw neural network values.
So we use **quantization**, which means:

real_value = int8_value × scaling_factor

Example:
Real weight = 0.123
INT8 value = 12
Scale factor = 0.01

Advantages:
✓ Ultra-fast
✓ Lowest memory
✓ Best for deployment
✓ Perfect for edge/Jetson devices
✓ Tensor Cores support INT8 matmul

Drawback:
• Reduces accuracy slightly
• Needs calibration dataset

──────────────────────────
Comparison Table

Format | Bits | Type | Speed | GPU Memory Use | Accuracy
FP32 | 32 | Floating | Slow | High | High accuracy
FP16 | 16 | Floating | Fast | Low | Good accuracy
BF16 | 16 | Floating | Fast | Low | Medium accuracy
INT8 | 8 | Integer | Fastest | Very Low | Lower accuracy without calibration

──────────────────────────
Why Do We Use Lower Precisions?

1. **Speed**
   FP16/INT8 can give 2× to 8× speed improvements.

2. **Memory**
   Half precision = half VRAM
   INT8 = 4× smaller than FP32

3. **Tensor Cores**
   Tensor Cores are optimized for:
   FP16
   BF16
   INT8

This is why TensorRT pushes FP16/INT8 for inference.

──────────────────────────
Where Each Precision Is Used

Training
• FP32 (classic)
• Mixed precision: FP16 + FP32
• BF16 (increasing trend)

Inference
• FP16 (best balance)
• INT8 (best speed)
• FP32 only if required for accuracy



## ONNX fundamentals: ONNX opsets, model graph, dynamic axes

Here is a **clear and detailed explanation** of ONNX fundamentals—**opsets**, **model graph**, and **dynamic axes**—exactly what you must know before converting models into TensorRT.

────────────────────────────
ONNX FUNDAMENTALS (Complete Explanation)

ONNX = *Open Neural Network Exchange*
It is a **standard format** for saving deep learning models so they can be shared across frameworks like:

• PyTorch → ONNX → TensorRT
• TensorFlow → ONNX → ONNX Runtime
• Scikit-Learn → ONNX → Edge deployment

ONNX stores your entire model in a `.onnx` file that contains:

• model architecture (layers)
• computational graph
• weights
• input/output shapes
• metadata

TensorRT uses ONNX as the most common method for importing models.

────────────────────────────

1. ONNX Opsets (Operator Sets)

An **opset** is a versioned set of operators (Conv, MatMul, ReLU, Add, etc.) that ONNX supports.

Think of opset as:
✔ The “version” of ONNX operations
✔ Defines how operators behave
✔ Ensures compatibility between frameworks and runtimes

Example:
`opset=11` has one set of supported ops
`opset=17` adds newer operators and improvements

Why Opsets Matter for TensorRT
TensorRT supports only certain opsets.
Example (for understanding):
• TensorRT 8.x supports opset 8–13
• TensorRT 10.x supports up to opset ~17 (varies)

If your ONNX model uses opset 20 but TensorRT supports only up to 17 → you will get errors like:

“Unsupported ONNX operator: XYZ”

So you must choose the opset wisely when exporting.

PyTorch export example:

```python
torch.onnx.export(model,
                  inputs,
                  "model.onnx",
                  opset_version=17)
```

────────────────────────────
2. ONNX Model Graph

An ONNX model is stored as a **computational graph**.

It contains:

2.1 Nodes
Each node = 1 operation (Conv, Matmul, Relu, Sigmoid)

Example:

```
Node 1: Conv
Node 2: BatchNorm
Node 3: Relu
Node 4: MaxPool
```

2.2 Edges
Edges are **tensors** that flow between nodes.

```
input → Conv → ReLU → output
```

2.3 Initializers
These are constant tensors (weights & biases).

2.4 Graph Input(s)
Defines input:
• name
• shape
• datatype

Example:

```
Input: [batch, 3, 224, 224]
```

2.5 Graph Output(s)
The final output tensor, e.g.:

```
Output: [batch, 1000]
```

You can visualize the graph using:

• Netron
• ONNX GraphSurgeon
• ONNX Explorer
• Netron.app (browser)

How TensorRT Uses the Graph
TensorRT parses the ONNX graph → optimizes it → fuses layers:

Conv + BatchNorm + ReLU
→ becomes
Single fused CUDA kernel

It removes unnecessary nodes, optimizes shapes, and selects best GPU kernels.

────────────────────────────
3. Dynamic Axes (Dynamic Shapes)

Dynamic axes define **flexible input shapes**.

Why Useful?
Sometimes input size is not fixed:

• variable batch size
• variable sequence length (NLP)
• variable image size
• variable video frames

Example use case
YOLO models accept images of any size
LLaMA models accept variable sequence lengths

Default ONNX is static:
batch = 1
image = 224x224

But you may want:
batch = 1–50
image size = 224–640
seq_length = 1–2048

Dynamic axes let you export ONNX with flexible dimensions.

PyTorch example:

```python
torch.onnx.export(
    model,
    inputs,
    "model.onnx",
    opset_version=17,
    dynamic_axes={
        "input": {0: "batch", 2: "height", 3: "width"},
        "output": {0: "batch"}
    }
)
```

This tells ONNX:

• axis 0 = dynamic batch
• axis 2 & 3 = dynamic height/width

TensorRT + Dynamic Shapes
TensorRT will require **optimization profiles**:

min shape → opt shape → max shape

Example for YOLO:

```
min: [1, 3, 320, 320]
opt: [4, 3, 640, 640]
max: [8, 3, 1280, 1280]
```

TensorRT builds different optimized engines for each shape range.

Advantages of Dynamic Shapes
✓ Allows real-time resizing
✓ Supports varied sequences in transformers
✓ Higher utilization in production
✓ No need to rebuild engine for each input size

Downside
• More memory usage
• Slightly lower optimization (than static)

────────────────────────────
Summary Table

Feature | Meaning | Why Important for TensorRT
Opset | Version of operators | Must match TRT compatibility
Graph | Network structure of nodes/tensors | TRT fuses & optimizes the graph
Dynamic Axes | Flexible input shapes | Needed for variable batch/image/sequence sizes

### Basic inference pipelines: batching, throughput vs latency

Here is a **clear, simple, yet deeply detailed** explanation of basic inference pipelines, batching, latency, and throughput — everything you must know before learning TensorRT or deploying optimized models.

────────────────────────────
BASIC INFERENCE PIPELINES (Foundation for TensorRT)

An **inference pipeline** is the full sequence of steps from receiving an input → running the model → returning output.

Typical steps:

1. Preprocessing
2. Moving data to GPU
3. Running the model
4. Postprocessing
5. Returning output

TensorRT speeds up Step 3 (model execution), but batching and throughput/latency decisions affect the entire pipeline.

Let's break down the key concepts.

────────────────────────────

1. Batching — What It Is and Why It Matters

**Batching means processing multiple inputs together in a single forward pass.**

Example:
Instead of running 1 image → run 8 images at the same time

Why batching helps?
Because GPUs are **massively parallel** and prefer larger workloads.

If you send:
• 1 image → GPU is underutilized
• 8 images → GPU is efficiently filled
• 32 images → maximum utilization (depending on memory)

Batching increases **throughput** but also increases **latency**.

────────────────────────────
2. Latency vs Throughput (Most Important Trade-Off)

Latency
= Time to process **one** request (single inference).

Throughput
= Number of inferences per second (overall system capacity).

They have an inverse relationship.

Example
• Batch size = 1
Latency = LOW
Throughput = LOW

• Batch size = 16
Latency = HIGH
Throughput = HIGH

This is why:

• Real-time applications → batch=1
• Offline processing → large batches
• Large-scale servers → dynamic batching

────────────────────────────
3. Latency Explained in Detail

Latency = Total time taken for producing the result.

Latency includes:
• Preprocessing latency
• GPU loading latency
• Model execution latency
• Postprocessing latency

Latency is measured as:
• p50 latency (median)
• p90 latency (slow 10%)
• p99 latency (slow 1%)

Applications requiring low latency:
• Self-driving cars
• Real-time video analytics
• Real-time chatbots
• Voice assistants
• Drones
• Industrial robots

Goal: Latency < 10ms or < 30ms

In TensorRT, FP16/INT8 greatly reduce latency.

────────────────────────────
4. Throughput Explained in Detail

Throughput = total number of samples processed per unit time.

Example:
• 200 images / second
• 500 queries / second
• 1000 tokens / second (LLM)

Applications requiring high throughput:
• Cloud inference servers
• Batch pipelines (nightly jobs)
• Video processing farms
• OCR batch jobs

Throughput increases when:
• batch size increases
• GPU utilization increases
• overlapping computation is enabled (streams)
• TensorRT engine is optimized
• quantization is applied

────────────────────────────
5. Why Batching Works (Under the Hood)

GPU has:
• thousands of CUDA cores
• Tensor Cores
• parallel SMs

A single image of 224×224×3 does not fully use GPU compute.

But batching:
• feeds a bigger tensor into GPU
• reduces kernel launch overhead
• maximizes SM occupancy
• better utilizes Tensor Cores
• reduces data transfer overhead

Example:
Batch=1 → GPU busy = 20%
Batch=32 → GPU busy = 95%

TensorRT uses batching to achieve peak performance.

────────────────────────────
6. Batch Size vs GPU Memory

Larger batch = more VRAM needed.

Example (ResNet50):
Batch=1 → 300 MB VRAM
Batch=16 → 800 MB
Batch=64 → 3 GB
Batch=128 → Out of memory

Your batch size is limited by:
• GPU VRAM
• Model size
• Precision (FP32 uses more VRAM than FP16)

FP16 reduces VRAM → supports larger batch sizes → boosts throughput.

────────────────────────────
7. Dynamic Batching (Used in Production Servers)

Dynamic batching means:

The server waits briefly (2–5 ms)
→ Collects multiple requests
→ Sends them together as a batch

This dramatically increases throughput without hurting latency too much.

Production systems like:
• Triton Inference Server
• Ray Serve
• TorchServe

All support dynamic batching.

Example:
Requests come:
[1], [1], [1], [1], [1]

Dynamic batcher merges them:
[5]

Huge speed improvement.

────────────────────────────
8. Streams and Overlapping in CUDA/TensorRT

Advanced inference pipelines overlap:
• data transfer
• kernel execution
• post-processing

Using cudaStream, you can overlap:
GPU can process batch N
while CPU prepares batch N+1

TensorRT also allows asynchronous inference:
context.execute_async_v2()

This boosts throughput further.

────────────────────────────
9. Real-World Example of Latency vs Throughput

Assume model is ResNet50.

Case 1: Real-time camera
Batch size = 1
Latency = 12 ms
Throughput = 80 FPS

Case 2: Offline dataset processing
Batch size = 32
Latency = 50 ms
Throughput = 640 FPS

Case 3: Server with dynamic batching
Batch size varies 1–16
Optimal balance between latency and throughput.

────────────────────────────
10. Summary Table

Concept | Meaning | Good For
Batching | Multiple inputs in one forward pass | High throughput
Latency | Time per inference | Real-time apps
Throughput | Inferences per second | Cloud/Batch processing
Dynamic Batching | Auto-merge requests | Production servers
Large Batch | High GPU usage | Offline pipelines
Small Batch | Low delay | Robotics, real-time



### Using NVIDIA GPU tools: nvidia-smi, CUDA toolkit, cuDNN

Here is a **clear and detailed explanation** of the three essential NVIDIA tools you must know before using TensorRT:
**nvidia-smi**, **CUDA Toolkit**, and **cuDNN**.

These form the foundation of GPU-based deep learning and model deployment.

────────────────────────────────────

1. NVIDIA-SMI (System Management Interface)

**nvidia-smi** is a command-line tool that comes with NVIDIA drivers.
It shows **real-time GPU usage, processes, temperature, power, memory, and driver details**.

Command:

```
$ nvidia-smi
```

You will see:

GPU | Name | Persistence-M | Bus-Id | Disp.A | Volatile Uncorr. ECC | Fan | Temp | Perf | Pwr | Usage | Memory-Usage | Processes

What you can monitor with nvidia-smi:

1.1 GPU Memory Usage
Shows how much VRAM your model uses:

```
|  GPU Memory Usage  |
|    350MiB / 12288MiB   |
```

1.2 GPU Utilization
Shows how busy your GPU is.

• low (0–20%) → you're underutilizing
• high (80–100%) → great utilization

1.3 Running Processes
Shows the apps currently using GPU:

```
python3  4500MiB
triton   2500MiB
```

1.4 GPU Temperature & Power
Useful for server deployments or overheating issues.

1.5 Kill a GPU process

```
sudo kill -9 <PID>
```

1.6 Monitoring

```
watch -n 1 nvidia-smi
```

This refreshes every second (good for watching inference load).

Why it is important in TensorRT
• To ensure TensorRT engine fits in VRAM
• To identify which processes are hogging GPU
• To verify GPU utilization before and after optimization
• To debug memory leaks

────────────────────────────────────
2. CUDA Toolkit (Core GPU Programming Library)

CUDA Toolkit is the **foundation** for all GPU computing.

It includes:

2.1 CUDA Compiler (nvcc)
Compiles CUDA kernels (.cu files).

2.2 CUDA Libraries
Core GPU libraries used by deep learning frameworks:
• cuBLAS (matrix multiplication)
• cuFFT (Fourier transforms)
• cuRAND (random numbers)
• cuSPARSE (sparse ops)

TensorRT uses these under the hood.

2.3 Runtime API & Driver API
Allows Python/C++ programs to run operations on GPU.

2.4 CUDA Samples
Example GPU programs – useful for learning GPU basics.

2.5 Device Query
Check GPU compatibility:

```bash
deviceQuery
```

This shows your GPU architecture (e.g., SM 89 for Ada GPUs).

2.6 nvcc
To check CUDA version:

```
nvcc --version
```

Why CUDA Toolkit matters for TensorRT
✓ TensorRT uses CUDA kernels internally
✓ FP16/INT8 layers require CUDA tensor core kernels
✓ Without CUDA, TensorRT cannot run at all
✓ Some TensorRT features require specific CUDA versions

Compatibility Rule
TensorRT version must match (or be compatible with):
CUDA version + cuDNN version + GPU architecture.

────────────────────────────────────
3. cuDNN (CUDA Deep Neural Network Library)

cuDNN = **CUDA library optimized for deep learning operations**.

────────────────────────────

## ✔ 1. TensorRT in One Sentence

**TensorRT is NVIDIA’s high-performance deep learning inference engine that optimizes your model to run extremely fast on GPUs (FP32 → FP16 → INT8) with minimal loss in accuracy.**

It is NOT a training framework.

It is a specialized **optimization + deployment** engine for inference.

────────────────────────────

## ✔ 2. Why TensorRT Exists (The Problem)

Training a model (PyTorch/TensorFlow) is slow but acceptable.
Inference, however, is done **millions of times** in production:

Examples:
• Real-time video (30–120 FPS)
• Self-driving car sensors
• Robotics
• Drones
• Chatbots
• LLMs running at scale
• Server-side inference for thousands of users

Raw PyTorch/TensorFlow inference is:
✗ not optimized
✗ slow
✗ memory-heavy
✗ not using Tensor Cores fully
✗ not suitable for high-throughput production

TensorRT fixes all of this.

────────────────────────────

## ✔ 3. What TensorRT Actually Does

TensorRT performs **8 categories of optimizations**:

### 1. **Layer Fusion**

Combines multiple operations into a single GPU kernel.

Example:
Conv → BatchNorm → ReLU
→ Fused into ONE optimized kernel
→ Dramatically faster

### 2. **Precision Optimization**

Reduces precision from FP32 → FP16 → INT8 (with calibration).

FP16 → ~2× faster
INT8 → ~4–8× faster

TensorRT automatically selects the best precision per layer.

### 3. **Kernel Auto-Tuning**

For each layer, TRT tries multiple algorithms and picks the fastest:
• cuDNN convolution algorithms
• Tensor Core kernels
• Memory layouts

This is called **tactic selection**.

### 4. **Dynamic Tensor Memory**

Efficient buffer allocation and reuse.
Reduces VRAM usage significantly.

### 5. **Dynamic Shapes Optimization**

Uses **optimization profiles** to support flexible input sizes:
• varying batch sizes
• varying image resolutions
• varying sequence lengths

Critical for NLP and YOLO models.

### 6. **Graph Optimizations**

• Removes redundant operations
• Reorders ops for speed
• Eliminates no-op layers
• Constant folding (precomputes constants)

### 7. **INT8 Quantization**

With calibration data, TensorRT creates an INT8 engine with minimal accuracy loss.

### 8. **Custom Plugins**

Lets you add your own GPU kernels for unsupported ops.

────────────────────────────

## ✔ 4. What TensorRT Is *Not*

TensorRT is **not**:
✗ a training library
✗ a deep learning framework
✗ a replacement for PyTorch/TensorFlow

TensorRT is **only for inference optimizations and deployment**.

## ✔ What Is a TensorRT Engine?

A **TensorRT Engine** (also called **plan file**) is a *fully optimized, hardware-specific, ready-to-run inference model* created by TensorRT.

It is the **output** of the TensorRT optimization process.

You train your model in PyTorch/TensorFlow → export ONNX → TensorRT → engine.

Example file:

```
model_fp16.engine
model_int8.engine
model.plan
```

This engine contains:
✓ optimized computation graph
✓ fused layers
✓ kernel tactics
✓ memory planning
✓ precision selections (FP16/INT8)
✓ supported dynamic shapes
✓ device-specific optimizations

The engine is like a **compiled binary** of your model that runs extremely fast on NVIDIA GPUs.

────────────────────────────────────────

## ✔ Why Is It Called “Serialized Plan File”?

Serialized → saved in binary format
Plan → execution plan for the GPU

TensorRT takes your model and creates a **plan** that defines:

• which kernels to run
• in what order
• with which precision
• with what memory layout
• using which Tensor Cores
• using which batch size
• using which optimization profile

That plan is then **serialized into a binary file** → `.engine`.

Later, the file can be:
• loaded by the TensorRT runtime
• used directly for inference

You cannot modify an engine — it is immutable.

Unlike ONNX, you don’t need to parse or build anything at runtime.

An engine created for:

will NOT be optimal for
• RTX 4090
• Jetson Orin

Often you must regenerate engines per hardware.

Runtime just feeds input → GPU executes → returns output.

For each layer, TensorRT chooses the best kernel from cuDNN, cuBLAS, etc.

### 3. **Precision Decisions**

Per-layer decision:
FP32? FP16? INT8?

Example:
Conv → FP16

Attention MatMul → INT8

### 4. **Memory Planning**

GPU engines reserve memory for:
• activations

• weights
• scratch buffers

TensorRT optimizes memory reuse — reducing VRAM footprint.

### 5. **Optimized Layer Ordering**

TensorRT may reorder operations for speed.

### 6. **Dynamic Shape Handling**

If your model supports dynamic shapes, engine contains **optimization profiles**.

### 7. **Calibrated INT8 Scales**

If INT8 calibration was used, the engine stores:
• scale factors
• quantization parameters

```python
with trt.Builder(logger) as builder, \
     builder.create_network() as network, \
     trt.OnnxParser(network, logger) as parser:

    parser.parse_from_file("model.onnx")
    config = builder.create_builder_config()
    config.set_flag(trt.BuilderFlag.FP16)

    engine = builder.build_serialized_network(network, config)
    open("model.engine", "wb").write(engine)
```

## Optimization Profiles

Here is a **clear, deeply detailed, production-level explanation** of **Optimization Profiles** in TensorRT — one of the MOST important features when dealing with **dynamic input shapes**, **YOLO models**, **Transformers**, and **LLMs**.

Optimization Profiles are essential for understanding how TensorRT builds fast, flexible engines.

────────────────────────────────────────

# ✅ What Are Optimization Profiles?

An **Optimization Profile** tells TensorRT:

> “My model will receive inputs of different shapes.
> Here are the minimum, optimum, and maximum sizes you must support.”

It is a **range of shapes** that TensorRT will optimize for.

Optimization profiles are used ONLY when inputs have **dynamic shapes**.

Without profiles → TensorRT CANNOT build dynamic-shape engines.

────────────────────────────────────────

# ✅ Why Do Optimization Profiles Exist?

Some models require flexible input sizes, such as:

✔ YOLO (variable image size: 320 → 640 → 1280)
✔ Batch size changes at runtime
✔ NLP models with variable sequence lengths (1 → 2048)
✔ Transformers with max_length, num_tokens changes
✔ Video models with varying frame sizes

TensorRT needs to know:

* What is the **minimum shape**?
* What shape will you use **most often** (optimal)?
* What is the **maximum shape**?

This allows TensorRT to:
✔ Pre-allocate safe memory buffers
✔ Pre-build optimized kernels for each range
✔ Avoid wasting GPU memory
✔ Ensure fast inference inside these bounds

────────────────────────────────────────

# ✅ When Do You Need Optimization Profiles?

ONLY when your model uses **dynamic axes** like:

• batch size = -1
• height = -1
• width = -1
• sequence length = -1
• number of tokens = -1

Example dynamic input:

```
(batch_size, 3, -1, -1)
```

TensorRT must know:

* What is the smallest height/width?
* What is the largest height/width?

────────────────────────────────────────

# ✅ Structure of an Optimization Profile

Each profile contains 3 shapes:

1. **MIN shape**
   → smallest shape engine must support
   → used to save memory

2. **OPT (optimal) shape**
   → the most common shape at runtime
   → TensorRT heavily optimizes for this one
   → best performance for OPT size

3. **MAX shape**
   → largest shape system must support
   → used to define memory limits

────────────────────────────────────────

# 🚀 Example: YOLOv8 Dynamic Shape Optimization Profile

YOLO input size can vary:

* 320×320
* 640×640
* 1280×1280

Define profile:

```python
profile = builder.create_optimization_profile()

profile.set_shape(
    "images",
    min=(1, 3, 320, 320),
    opt=(4, 3, 640, 640),
    max=(8, 3, 1280, 1280)
)

config.add_optimization_profile(profile)
```

What this tells TensorRT:
• Minimum input: batch=1, size=320
• Optimize best for: batch=4, 640
• Maximum allowed: batch=8, 1280

────────────────────────────────────────

# 🚀 Example: LLM Optimization Profile (Transformer Models)

LLMs have dynamic sequence length:
1 → 2048 tokens

```python
profile.set_shape(
    "input_ids",
    min=(1, 1),
    opt=(1, 512),
    max=(1, 2048)
)
```

TensorRT uses:
• smallest KV-cache for min
• best speed for opt
• longest cache for max

────────────────────────────────────────

# 🚀 Example: Dynamic Batch Size

If you want batch = 1–32:

```python
profile.set_shape(
    "input",
    min=(1, 3, 224, 224),
    opt=(16, 3, 224, 224),
    max=(32, 3, 224, 224)
)
```

────────────────────────────────────────

# 🚀 Multiple Profiles (Important!)

A single engine can contain **multiple optimization profiles**.

Why?
In production (Triton, servers), batch sizes vary frequently.

Example with 3 profiles:

* Profile 0: [batch 1–4]
* Profile 1: [batch 5–16]
* Profile 2: [batch 16–64]

This allows:
✔ maximum performance
✔ minimal latency
✔ optimal kernel choice per range

Each profile has separate tactics optimized for its size range.

────────────────────────────────────────

# 🚀 How Optimization Profiles Affect Performance

TensorRT builds kernels for:
• min shape
• opt shape
• max shape

Most optimized kernels = **opt shape**
Worst performance = **max shape**
Medium = any shape between min and max

So:

### If you ALWAYS use one input shape (example: 640×640)

→ Set min = opt = max = (1, 3, 640, 640)

This produces **max performance**.

### If you need flexibility

→ Add realistic ranges
→ But performance will vary per shape

────────────────────────────────────────

# 🚀 What Happens If Input Exceeds Optimization Profile Range?

If your input shape is outside min/max:

TensorRT will THROW AN ERROR.

Example:
You defined max = (8, 3, 1280, 1280)
But you send (1, 3, 1500, 1500)

Error:

```
shape exceeds maximum allowed dimension
```

Solution:
Increase max shape or add a new optimization profile.

────────────────────────────────────────

# 🚀 Adding Optimization Profiles (Python Example)

```python
config = builder.create_builder_config()

profile = builder.create_optimization_profile()
profile.set_shape("input", min_shape, opt_shape, max_shape)

config.add_optimization_profile(profile)
```

Add more profiles:

```python
config.add_optimization_profile(profile1)
config.add_optimization_profile(profile2)
```

────────────────────────────────────────

# 🚀 Using Optimization Profiles in Runtime

When loading an engine:

```python
context = engine.create_execution_context()
context.set_optimization_profile(profile_index)
```

TensorRT selects kernels optimized for that profile.

────────────────────────────────────────

# 🚀 Optimization Profiles and Memory Planning

The max shape determines:
• maximum GPU memory reserved
• scratch buffer sizes
• workspace size needed

Bigger max shape = more memory needed.

This is why:
YOLOv8 1280×1280 max shape uses MUCH more VRAM than YOLO 640×640.

────────────────────────────────────────

Opt Shape | Most optimized shape (fastest)
Max Shape | Highest allowed size (memory bound)

Multiple Profiles | Better latency + throughput
Fail Cases | Input outside min–max → ERROR

Optimization profiles are essential for:

## Precision Modes (FP32 / FP16 / INT8)

Here is a **clear, deeply detailed, technical-but-simple explanation** of **Precision Modes in TensorRT (FP32, FP16, INT8)** — how they work, why they matter, and how TensorRT uses them to accelerate inference.

This is one of the most important topics in TensorRT.

────────────────────────────────────────

# ✅ What Are Precision Modes?

Precision modes define the **numeric format** used for tensors (weights + activations) during inference.

TensorRT supports:

### ✔ FP32 → Full precision

### ✔ FP16 → Half precision (Tensor Core accelerated)

### ✔ INT8 → Integer precision (quantized, fastest)

Each precision impacts:

• **speed**
• **accuracy**
• **memory usage**
• **GPU requirements**

TensorRT mixes these modes layer-by-layer for best performance with minimal accuracy loss.

────────────────────────────────────────

# 1️⃣ FP32 (Single Precision, 32-bit)

### ✔ 32 bits per number

### ✔ Most accurate

### ✔ Slowest

### ✔ Uses most memory

### ✔ Does NOT exploit Tensor Cores

FP32 layout:

* 1 sign bit
* 8 exponent bits
* 23 mantissa bits

Used when:

• Accuracy is critical
• FP16/INT8 unstable
• Certain layers need high numeric range

But in modern GPUs, FP32 is the slowest inference mode.

### How to enable FP32 (default):

No flag needed:

```bash
trtexec --onnx=model.onnx
```

────────────────────────────────────────

# 2️⃣ FP16 (Half Precision, 16-bit)

### ✔ 16 bits → half memory

### ✔ Much faster (2–4×)

### ✔ Uses Tensor Cores

### ✔ Accuracy usually very close to FP32

### ✔ Best performance/accuracy balance

FP16 format:

* 1 sign bit
* 5 exponent bits
* 10 mantissa bits

FP16 reduces precision slightly, but neural networks tolerate it extremely well.

TensorRT selects FP16 kernels automatically when supported.

### How to enable FP16:

```bash
trtexec --onnx=model.onnx --fp16
```

Python:

```python
config.set_flag(trt.BuilderFlag.FP16)
```

When FP16 is used:
• conv layers
• matmul layers
• attention layers
• normalization
• activations

Tensor Cores give HUGE acceleration to FP16.

### Example: Speed Gain

ResNet50:
FP32 → 100 ms
FP16 → 35 ms

YOLOv8:
FP32 → 40 ms
FP16 → 12 ms

────────────────────────────────────────

# 3️⃣ INT8 (8-bit Integer Precision)

### ✔ Fastest (2–4× faster than FP16)

### ✔ Lowest memory (4× smaller vs FP32)

### ✔ Uses Tensor Cores

### ✔ Requires calibration dataset

### ✔ Slight accuracy drop (can be minimized)

INT8 stores values as:

```
int_value × scale_factor
```

TensorRT INT8 quantization requires:
• Calibration data (representative samples)
• Calibration cache generation

Two main calibrators:

* EntropyCalibrator2
* MinMaxCalibrator

Using INT8 gives the **highest throughput** for:

• YOLO
• Classification models
• LLMs with INT8 weights
• Real-time video inference
• Jetson devices

### How to enable INT8:

```bash
trtexec --onnx=model.onnx --int8 --calib=calib.cache
```

Python:

```python
config.set_flag(trt.BuilderFlag.INT8)
config.int8_calibrator = MyCalibrator()
```

### Accuracy

INT8 accuracy drop:
Usually < 1% if calibrated well.

────────────────────────────────────────

# 🔥 TensorRT Uses Mixed Precision Per Layer

TensorRT does NOT force one precision for the entire model.

It makes smart decisions:

Layer | Precision
Conv | FP16 or INT8
MatMul | FP16 or INT8 (Tensor Core)
LayerNorm | FP32 (for stability)
Softmax | FP32 (for stability)
Residual Add | FP16
GELU | FP32 or FP16

This ensures:
✔ high accuracy
✔ maximum performance

This is called **precision calibration** inside TensorRT.

────────────────────────────────────────

# 🔥 VRAM Usage Comparison

Format | Memory Required | Speed
FP32 | 100% | Slow
FP16 | 50% | Fast
INT8 | 25% | Fastest

If you're deploying large LLMs:
INT8 is essential for fitting in VRAM.

────────────────────────────────────────

# 🔥 GPU Hardware Requirements

Precision | Requires Tensor Cores?
FP32 | No
FP16 | Yes (for acceleration)
INT8 | Yes (for acceleration)

Supported GPUs:

* RTX 20xx
* RTX 30xx
* RTX 40xx
* Tesla T4 / V100 / A100 / H100
* Jetson Xavier / Orin

Jetson Nano supports only FP16 (no INT8 Tensor Cores).

────────────────────────────────────────

# 🔥 Which Precision Should You Use?

### ✔ For real-time inference

Use **FP16** or **INT8**

### ✔ For maximum throughput

Use **INT8**

### ✔ For highest accuracy

Use **FP32** (rare in deployment)

### ✔ For edge devices (Jetson)

Use **INT8**

### ✔ For LLMs

Use **FP16**, **BF16**, or **INT8**
(LLM INT8 uses weight-only quantization)

## Execution Context

Here is a **clear, deep, and practical explanation** of **Execution Context** in TensorRT — one of the MOST important concepts for running high-performance inference.

Execution Context is the runtime engine that actually executes your TensorRT engine on the GPU.

────────────────────────────────────────

# ✅ What Is an Execution Context?

A **TensorRT Execution Context** is the *runtime instance* that executes the optimized TensorRT engine with a specific input shape and optimization profile.

In simple words:

> **The TensorRT Engine is the compiled model (.engine file).
> The Execution Context is the active runtime object that runs inference using that engine.**

It handles:
✔ input shapes
✔ memory bindings
✔ workspace buffers
✔ optimization profiles
✔ batch sizes
✔ dynamic shape updates
✔ running inference asynchronously

You **must create an Execution Context** before running inference.

────────────────────────────────────────

# ✅ Why Execution Context Exists?

TensorRT engines contain reusable kernels and memory plans, but:

❌ They cannot execute by themselves
❌ They cannot hold dynamic shapes
❌ They cannot manage per-inference buffers

Execution Context solves this by:

✔ binding input/output tensors
✔ selecting the correct optimization profile
✔ allocating workspace memory
✔ running inference

Each context is lightweight and reusable.

────────────────────────────────────────

# 🧠 Key Concept:

**One engine can have multiple execution contexts running in parallel**
to maximize GPU throughput.

Used in:
• Triton Inference Server
• Multi-stream video pipelines
• Real-time multi-user LLMs

────────────────────────────────────────

# 🚀 Execution Context Responsibilities (Deep Explanation)

## 1️⃣ Select Optimization Profile

If the engine contains multiple profiles:

```python
context.set_optimization_profile(0)
```

The context loads memory layouts optimized for that shape range.

## 2️⃣ Set Input Shapes (for dynamic engines)

Example:

```python
context.set_binding_shape(0, (1, 3, 640, 640))
```

TensorRT validates that:
✔ shape is within min–max
✔ memory fits
✔ correct kernel tactic exists

If shape is invalid, inference fails.

## 3️⃣ Bind Input/Output Memory Buffers

You must allocate GPU buffers and bind them:

```python
context.execute_v2(bindings)
```

Bindings point to:
• input GPU memory
• output GPU memory

TensorRT uses these to run inference.

## 4️⃣ Manage Workspace Memory

Context allocates temporary GPU memory for:
• intermediate activations
• scratch buffers
• tensor reshaping
• layer execution

This depends on:
• precision mode
• shape
• optimization profile

## 5️⃣ Run Actual Inference

Two ways:

### Synchronous:

```python
context.execute_v2(bindings)
```

### Asynchronous:

```python
context.execute_async_v2(bindings, stream_handle)
```

Asynchronous = faster because it overlaps:
• CPU preprocessing
• GPU compute
• CPU post-processing

# ✅ 1. TensorRT Plugins (Custom Layers)

## ✔ What is a Plugin in TensorRT?

A **TensorRT Plugin** is a **custom GPU layer** that you implement yourself when:

✔ A layer in your model is not supported by TensorRT  
✔ You want to replace a slow kernel with your own optimized kernel  
✔ You want to implement custom model logic (YOLO decode head, custom activation, etc.)  
✔ An ONNX op cannot be parsed into TensorRT  
✔ You want to fuse multiple layers into one custom GPU kernel  

Plugins allow TensorRT to execute ANY custom or unsupported operation.

---

## ✔ Why Plugins Exist?

Not all deep learning ops are supported by TensorRT’s built-in layers.

Examples:

● YOLOv3/v4/v5/v7/v8 decode layers  
● NMS (Non-Maximum Suppression) before TRT 8  
● Custom attention mechanisms  
● Swish / Mish activation (older TRT versions)  
● Special mathematical operations  
● LLM fused ops (FlashAttention-like kernels)  

When the ONNX parser hits an unsupported op like this:
```
ERROR: Unsupported ONNX operator: GridSampler
```

→ You must write a plugin.

---

## ✔ Plugin Types

### 🔸 1. **IPluginV2**
Older API for static shape layers.

### 🔸 2. **IPluginV2DynamicExt** (Most used)
Supports:
- Dynamic shapes  
- FP16  
- INT8  
- Per-tensor formats  

This is used for modern models.

---

## ✔ What a Plugin Contains

You must implement:

1. **getOutputDimensions**  
   Defines output tensor shape.

2. **supportsFormatCombination**  
   Tells TensorRT which precisions (FP32/FP16/INT8) the plugin supports.

3. **enqueue()**  
   The HEART of the plugin.  
   Runs the CUDA kernel (your GPU code).

4. **serialize() / deserialize()**  
   Save & load plugin state inside engine.

5. **clone()**  
   How TensorRT copies the plugin for parallel contexts.

---

## ✔ Typical Use Cases

### YOLO models  
Implement grid decoding + post-processing as a plugin.

### LLMs / Transformers  
Fused:
- QKV projection  
- Softmax  
- LayerNorm  
- Rotary embeddings  

### Robotics  
Custom image processing kernels.

### Scientific models  
Special math ops not available in ONNX.

---

## ✔ Plugin Workflow (High-Level)

1. Write plugin (C++ or Python/CUDA)  
2. Register plugin creator  
3. ONNX parser detects custom op  
4. TensorRT inserts plugin into network  
5. Engine building proceeds normally  
6. During inference, TRT calls your plugin’s CUDA kernel  

---

## ✔ Tools for Working With Plugins

• TensorRT Plugin Registry  
• ONNX GraphSurgeon  
• Polygraphy  
• Custom C++/CUDA kernels  

Plugins are **advanced**, but extremely powerful.

# ✅ 2. INT8 Calibration in TensorRT (Entropy / MinMax / SQ)

TensorRT supports **INT8 precision**, which is 2–4× faster than FP16.

But because INT8 is low precision, TensorRT needs to know:
> “What numeric range do your tensors usually fall into?”

This is learned using **calibration**.

---

# 🎯 Why Calibration?

INT8 stores values in 8 bits:
```
int8_value = real_value / scale_factor
```

TensorRT must compute the correct **scale_factor** to avoid:
✔ overflow  
✔ underflow  
✔ major accuracy loss  

This is done using a calibration dataset.

---

# 🔥 Two Main Calibrator Types in TensorRT

## 1️⃣ **EntropyCalibrator2**  
✔ Most common  
✔ Best for classification & detection  
✔ Uses KL divergence to pick quantization ranges  
✔ Good accuracy retention  

Entropy = “distribution matching”

TensorRT tries to match FP32 distribution → INT8 distribution.

### Usage:
```python
config.set_flag(trt.BuilderFlag.INT8)
config.int8_calibrator = trt.IInt8EntropyCalibrator2(...)
```

---

## 2️⃣ **MinMaxCalibrator**
✔ Fastest calibration  
✔ Uses min and max values directly  
✔ Good for symmetric distributions  
✔ May lose accuracy on complex models  
✔ Used for simpler CNNs or industrial pipelines

### Usage:
```python
config.int8_calibrator = trt.IInt8MinMaxCalibrator(...)
```

---

## 3️⃣ **SmoothQuant (SQ) — For LLMs / Transformers**
This is the new INT8 method for LLMs.

✔ Smooths out activation spikes  
✔ Best accuracy for Transformers  
✔ Used in TensorRT-LLM  
✔ Extremely stable for GPT, LLaMA, BERT, T5  

### Benefits:
- Removes quantization outliers  
- Works without huge accuracy loss  
- Faster inference with 8-bit weights  

---

# 🎯 How INT8 Calibration Works (Step-by-Step)

1. Provide a **representative dataset**  
   (like 100–500 images or text samples)

2. TensorRT runs the model internally using FP32  
   → Collects activation statistics

3. Computes **scaling factors**  
   using Entropy or MinMax methods

4. Generates **calibration cache file**
```
calib.cache
```

5. During engine building TensorRT loads:
```
scale factors → quantizes layers → builds INT8 engine
```

---

# 🎯 Calibration Dataset Guidelines

✔ 50–500 samples  
✔ Realistic distribution  
✔ Same preprocessing as real inference  
✔ No labels required  
✔ No need for large dataset  

Example datasets:
- ImageNet small subset  
- 100 frames from a camera  
- 200 random JPGs  
- 100 sample prompts for LLMs  

---

# 🎯 Implementing Your Own Calibrator (Python)

TensorRT requires you to subclass:

### For Entropy:
```python
class MyCalibrator(trt.IInt8EntropyCalibrator2):
    def get_batch(self): ...
    def read_calibration_cache(self): ...
    def write_calibration_cache(self, cache): ...
```

### For MinMax:
```python
class MyCalibrator(trt.IInt8MinMaxCalibrator):
    ...
```

Supplies batches to TensorRT until calibration is complete.
